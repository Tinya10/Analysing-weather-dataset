```python
#Generate a pivot table to contain aggregate values (like mean, max, min, sum, len) recorded by a month
def agg_values_ina_month(df,date_col,agg_col, agg):
    
    df[date_col] = pd.to_datetime(df[date_col])
    aggregate = {'mean':np.mean,'max':np.max,'min':np.min,'sum':np.sum,'len':len}
    aggregated_value = df.pivot_table(values=[agg_col], index=df[date_col].dt.month,aggfunc={agg_col:aggregate[agg]})
    return aggregated_value

agg_values_ina_month(weather,'Date/Time','Dew Point Temp (C)','mean')
```

```python
#Pivot table without creating a function
agg_values_ina_month = weather.pivot_table(values=['Dew Point Temp (C)'], index = weather['Date/Time'].dt.month,aggfunc= np.mean)
agg_values_ina_month
```

```python
#Group by with the help of function to calculate mean values for weather

def group_values(df,col1,agg1):
    aggregate = {'mean':np.mean,'max':np.max,'min':np.min,'sum':np.sum,'len':len}
    grouping = df.groupby(col1).agg(aggregate[agg1])
    return grouping
mean_weather = group_values(weather,"Weather",'mean')
mean_weather
```

```python
#Group by to calculate mean values for weather without creating a function

mean_weather = weather.groupby(['Weather']).agg([np.mean])
mean_weather

```

```python
#convert Celsius temperature into Fahrenheit temperatures with the help of function

def convert(df,celsius):
    centigrade_temps = df[celsius]
    converted_temp =  1.8*centigrade_temps + 32
    return converted_temp

weather_fahrehheit = convert(weather,"Temp (C)")
weather_fahrehheit
```

```python
#convert Celsius temperature into Fahrenheit temperatures without function

weather_fahrehheit = 1.8*weather['Temp (C)']+32
weather_fahrehheit
```
